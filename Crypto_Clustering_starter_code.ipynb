```python
# Imported required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Loaded the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Displayed sample data
df_market_data.head(10)
```

```python
# Generated summary statistics
df_market_data.describe()
```

```python
# Plotted the data in the DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Used the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()

market_data_scaled = scaler.fit_transform(df_market_data)
market_data_scaled[:5]
```

```python
# Created a DataFrame with the scaled data
df_market_scaled = pd.DataFrame(market_data_scaled)

# Copied the crypto names from the original DataFrame
scaled_df = df_market_data.index.copy()

# Set the coin_id column as index
df_market_scaled.index = scaled_df

# Displayed the scaled DataFrame
df_market_scaled.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Created a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Created an empty list to store the inertia values
inertia = []

# Created a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Created a KMeans model using the loop counter for the n_clusters
# 2. Fitted the model to the data using `df_market_data_scaled`
# 3. Appended the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_market_scaled)
    inertia.append(k_model.inertia_)
```

```python
# Created a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plotted a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** When K = 4

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialized the K-Means model using the best value for k
kmeans_model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fitted the K-Means model using the scaled DataFrame
kmeans_model.fit(df_market_scaled)
```

```python
# Predicted the clusters to group the cryptocurrencies using the scaled DataFrame
cluster_predictions = kmeans_model.predict(df_market_scaled)

# Displayed the results from the array of cluster values.
cluster_predictions
```

```python
# Created a copy of the scaled DataFrame
df_market_data_copy = df_market_data.copy()
```

```python
# Added a new column to the copy of the scaled DataFrame with the predicted clusters
df_market_data_copy["Cluster"] = cluster_predictions

# Displayed the copy of the scaled DataFrame
df_market_data_copy.head()

```

```python
# Created a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Colored the graph points with the labels found using K-Means and
# added the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
cluster = df_market_data_copy.hvplot.scatter(
    x="price_change_percentage_24h", 
    y="price_change_percentage_7d",
    by="Cluster",  
    hover_cols=["coin_id"]
)
cluster
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Created a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Used the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
scaled_pca = pca.fit_transform(df_market_scaled)

# Viewed the scaled PCA data
scaled_pca[:5]
```

```python
# Retrieved the explained variance to determine how much information
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** About 89.5% of the total variance is condensed into the 3 PCA variables.

```python
# Created a new DataFrame with the PCA data.
scaled_pca_df = pd.DataFrame(
    scaled_pca,
    columns=["PCA1", "PCA2", "PCA3"]
)
# Copied the crypto names from the original scaled DataFrame
scaled_pca_df["coin_id"] = df_market_scaled.index.copy()

# Set the coin_id column as index
scaled_pca_df.set_index("coin_id", inplace=True)

# Displayed the scaled PCA DataFrame
scaled_pca_df.head()
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Created a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Created an empty list to store the inertia values
inertia = []

# Created a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Created a KMeans model using the loop counter for the n_clusters
# 2. Fitted the model to the data using `df_market_data_pca`
# 3. Appended the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=0)
    k_model.fit(scaled_pca_df)
    inertia.append(k_model.inertia_)
```

```python
# Created a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}
# Created a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
# Plotted a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow = df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
elbow
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** When K=4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No difference.

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialized the K-Means model using the best value for k
kmeans = KMeans(n_clusters=4, random_state=0)
```

```python
# Fitted the K-Means model using the PCA data
kmeans.fit(scaled_pca_df)
```

```python
# Predicted the clusters to group the cryptocurrencies using the scaled PCA DataFrame
cluster_labels = kmeans.predict(scaled_pca_df)

# Printed the resulting array of cluster values.
scaled_pca_df.head()
```

```python
# Created a copy of the scaled PCA DataFrame
scaled_pca_df_copy = scaled_pca_df.copy()

# Added a new column to the copy of the PCA DataFrame with the predicted clusters
scaled_pca_df_copy["Cluster"] = kmeans.labels_

# Displayed the copy of the scaled PCA DataFrame
scaled_pca_df_copy.head()
```

```python
# Created a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Colored the graph points with the labels found using K-Means and
# added the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
scaled = scaled_pca_df_copy.hvplot.scatter(
    x="PCA1", 
    y="PCA2",
    by="Cluster",  
    hover_cols=["coin_id"]
)

scaled
```
